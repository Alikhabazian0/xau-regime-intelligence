# Adding more features

```python
import pandas as pd
import pandas_datareader as pdr
from datetime import datetime
```

```python
# 1. Load raw data
df_xau = pd.read_csv(r"C:\\Users\\user\\OneDrive\\Desktop\\financial_regime_clasification\\data\\raw\\raw.csv", parse_dates=['date'])
df_xau.set_index('date', inplace=True)
```

```python
# 2. date range from data
start_date = df_xau.index.min()
end_date = df_xau.index.max()

print(f"XAUUSD data: {len(df_xau)} rows from {start_date} to {end_date}")
```

```python
# 3. Download VIX and DXY
try:
    vix_data = pdr.DataReader('VIXCLS', 'fred', start_date, end_date)
    vix_data.rename(columns={'VIXCLS': 'vix'}, inplace=True)
    
    dxy_data = pdr.DataReader('DTWEXBGS', 'fred', start_date, end_date)
    dxy_data.rename(columns={'DTWEXBGS': 'dxy'}, inplace=True)
    
    print(f" Downloaded VIX: {len(vix_data)} rows")
    print(f" Downloaded DXY: {len(dxy_data)} rows")
    
except Exception as e:
    print(f"Download error: {e}")
    vix_data = pd.DataFrame()
    dxy_data = pd.DataFrame()
```

```python
# 4. Merge with raw data
combined_df = df_xau.copy()
combined_df = combined_df.join(vix_data, how='left')
combined_df = combined_df.join(dxy_data, how='left')
```

```python
# 5. Forward fill missing values
combined_df[['vix', 'dxy']] = combined_df[['vix', 'dxy']].fillna(method='ffill')
```

```python
# 6. Save to processed folder
combined_df.to_csv(r"C:\\Users\\user\\OneDrive\Desktop\\financial_regime_clasification\\data\\processed\\xau_with_vix_dxy.csv")

print(f"\n Saved to: data/processed/xau_with_vix_dxy.csv")
print(f"Shape: {combined_df.shape}")
print(f"\nFirst 5 rows:")
print(combined_df[['close', 'vix', 'dxy']].head())
```
