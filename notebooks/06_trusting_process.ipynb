```python
import pandas as pd

df = pd.read_csv("C:\\Users\\user\\OneDrive\\Desktop\\financial_regime_clasification\\data\\raw\\XAUUSD_5min_all.csv")

df['datetime'] = pd.to_datetime(
    df['Date'].astype(str) + " " + df['Timestamp']
)

df = df.set_index('datetime')

df_1h = (
    df.resample("1h")
    .agg({
        "Open": 'first',
        'High': 'max',
        'Low': 'min',
        'Close': 'last',
        "Volume": 'sum'
    })
)

df_1h = df_1h.dropna()

df_1h.columns = [
    'open', 
    'high',
    'low',
    'close',
    'volume'
]

df_1h = df_1h.reset_index()
df_1h.rename(columns={'datetime': 'date'}, inplace=True)

print(df_1h.head())
print(df_1h.shape)
```

```python
df_1h.to_csv(
    "C:\\Users\\user\\OneDrive\\Desktop\\financial_regime_clasification\\data\\raw\\xauusd_1h_2020_2025.csv",
    index=False
)
```
